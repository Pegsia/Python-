# äºŒã€è‡ªå®šä¹‰ç½‘ç»œå±‚
  

å¯¹äºè‡ªå®šä¹‰çš„ç½‘ç»œå±‚ï¼Œéœ€è¦å®ç°åˆå§‹åŒ–__init__æ–¹æ³•å’Œå‰å‘ä¼ æ’­é€»è¾‘ call æ–¹æ³•ã€‚æˆ‘ä»¬ä»¥æŸä¸ªå…·ä½“çš„è‡ªå®šä¹‰ç½‘ç»œå±‚ä¸ºä¾‹ï¼Œå‡è®¾æˆ‘ä»¬éœ€è¦ä¸€ä¸ªæ²¡æœ‰åç½®çš„å…¨è¿æ¥å±‚ï¼Œå³ bias ä¸º 0ï¼ŒåŒ æ—¶å›ºå®šæ¿€æ´»å‡½æ•°ä¸º ReLU å‡½æ•°ã€‚å°½ç®¡è¿™å¯ä»¥é€šè¿‡æ ‡å‡†çš„ Dense å±‚åˆ›å»ºï¼Œä½†æˆ‘ä»¬è¿˜æ˜¯å®ç°è¿™ä¸ªè‡ªå®šä¹‰ç±»ã€‚

é¦–å…ˆåˆ›å»ºç±»å¹¶ç»§æ‰¿è‡ª Layer åŸºç±»ï¼Œåˆ›å»ºåˆå§‹åŒ–æ–¹æ³•ï¼Œå¹¶è°ƒç”¨æ¯ç±»çš„åˆå§‹åŒ–å‡½æ•°ï¼Œç”±äº æ˜¯å…¨è¿æ¥å±‚ï¼Œæˆ‘ä»¬éœ€è¦è®¾ç½®ç‰¹å¾çš„é•¿åº¦ inp_dim å’Œè¾“å‡ºç‰¹å¾çš„é•¿åº¦ outp_dimï¼Œå¹¶é€šè¿‡self.add_variable(name, shape)åˆ›å»º shape å¤§å°ï¼Œåå­—ä¸º name çš„å¼ é‡ï¼Œå¹¶è®¾ç½®ä¸ºéœ€è¦ä¼˜åŒ–:

```python
from tensorflow.keras import layers
class MyDense(layers.Layer): # è‡ªå®šä¹‰ç½‘ç»œå±‚
    def __init__(self, inp_dim, outp_dim):
        super(MyDense, self).__init__()
    # åˆ›å»ºæƒå€¼å¼ é‡å¹¶æ·»åŠ åˆ°ç±»ç®¡ç†åˆ—è¡¨ä¸­ï¼Œè®¾ç½®ä¸ºéœ€è¦ä¼˜åŒ–
        self.kernel = self.add_variable('w', [inp_dim, outp_dim],trainable=True)
net = MyDense(4,3) # åˆ›å»ºè¾“å…¥ä¸º 4ï¼Œè¾“å‡ºä¸º 3 èŠ‚ç‚¹çš„è‡ªå®šä¹‰å±‚
net.variables
```

# ä¸‰ã€è‡ªå®šä¹‰ç½‘ç»œ

    åœ¨å®Œæˆäº†æˆ‘ä»¬è‡ªå®šä¹‰çš„å…¨è¿æ¥å±‚ç±»ä¹‹åï¼Œæˆ‘ä»¬åŸºäºä¸Šè¿°çš„â€œæ— åç½®çš„å…¨è¿æ¥å±‚â€æ¥å®ç° MNIST æ‰‹å†™æ•°å­—å›¾ç‰‡æ¨¡å‹çš„åˆ›å»ºã€‚è‡ªå®šä¹‰çš„ç±»å¯ä»¥å’Œå…¶ä»–æ ‡å‡†ç±»ä¸€æ ·ï¼Œé€šè¿‡ Sequential å®¹å™¨æ–¹ä¾¿åœ°åŒ…è£¹æˆä¸€ä¸ªç½‘ç»œæ¨¡å‹:

```python
import tensorflow as tf
network =tf.keras.models.Sequential([MyDense(784, 256), # ä½¿ç”¨è‡ªå®šä¹‰çš„å±‚ MyDense(256, 128),
                 MyDense(128, 64),
                 MyDense(64, 32),
                 MyDense(32, 10)])
network.build(input_shape=(None, 28*28)) 
```

```python
network.summary()
```

     å¯ä»¥çœ‹åˆ°ï¼Œé€šè¿‡å †å æˆ‘ä»¬çš„è‡ªå®šä¹‰ç±»ï¼Œä¸€æ ·å¯ä»¥å®ç° 5 å±‚çš„å…¨è¿æ¥å±‚ç½‘ç»œï¼Œæ¯å±‚å…¨è¿æ¥å±‚ æ— åç½®å¼ é‡ï¼ŒåŒæ—¶æ¿€æ´»å‡½æ•°å›ºå®šä½¿ç”¨ ReLUã€‚æ›´æ™®éåœ°ï¼Œæˆ‘ä»¬å¯ä»¥ç»§æ‰¿åŸºç±»æ¥å®ç°ä»»æ„é€»è¾‘çš„è‡ªå®šä¹‰ç½‘ç»œç±»ã€‚ä¸‹é¢æˆ‘ä»¬æ¥åˆ›å»ºè‡ªå®š ä¹‰ç½‘ç»œç±»ï¼Œé¦–å…ˆåˆ›å»ºå¹¶ç»§æ‰¿ Model åŸºç±»ï¼Œåˆ†å¸ƒåˆ›å»ºå¯¹åº”çš„ç½‘ç»œå±‚å¯¹è±¡:

```python
from tensorflow import keras
class MyModel(keras.Model):
    # è‡ªå®šä¹‰ç½‘ç»œç±»ï¼Œç»§æ‰¿è‡ª Model åŸºç±» 
    def __init__(self):
        super(MyModel, self).__init__() # å®Œæˆç½‘ç»œå†…éœ€è¦çš„ç½‘ç»œå±‚çš„åˆ›å»ºå·¥ä½œ self.fc1 = MyDense(28*28, 256)
        self.fc2 = MyDense(256, 128)
        self.fc3 = MyDense(128, 64) 
        self.fc4 = MyDense(64, 32)
        self.fc5 = MyDense(32, 10)
        #ç„¶åå®ç°è‡ªå®šä¹‰ç½‘ç»œçš„å‰å‘è¿ç®—é€»è¾‘:
    def call(self, inputs, training=None): # è‡ªå®šä¹‰å‰å‘è¿ç®—é€»è¾‘
        x = self.fc1(inputs)
        x = self.fc2(x)
        x = self.fc3(x)
        x = self.fc4(x) 
        x = self.fc5(x)
        return x
```
